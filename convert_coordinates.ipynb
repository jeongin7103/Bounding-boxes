## 상대좌표(w,h,x,y) --> 절대좌표(x_min, y_min, x_max, y_max) 로 변환해주는 함수

```python
def convert(json_dir):

    for filename in os.listdir(json_dir):
        if filename.endswith(".json"):  

            json_file_path = os.path.join(json_dir, filename)  

            with open(json_file_path, 'r') as json_file:
                data = json.load(json_file)

            for item in data:
                w = float(item["W"])
                h = float(item["H"])
                x, y = map(float, item["Point(x,y)"].split(','))

                resized_width = 640
                resized_height = 640

                x_min = (x - w / 2) * resized_width
                y_min = (y - h / 2) * resized_height
                x_max = (x + w / 2) * resized_width
                y_max = (y + h / 2) * resized_height

                new_xmin = x_min
                new_ymin = y_min
                new_xmax = x_max
                new_ymax = y_max

                print(new_xmin, new_ymin, new_xmax, new_ymax)

                item["boxes"] = [new_xmin, new_ymin, new_xmax, new_ymax]

            with open(json_file_path, 'w') as updated_json_file:
                json.dump(data, updated_json_file, indent=4)

            # 업데이트된 JSON 파일 경로 출력
            print(f"업데이트된 JSON 파일: {json_file_path}")
```

## 변환할 json 파일 경로 넣어주기

```python
# JSON 파일이 들어 있는 디렉토리 경로
json_dir = "/Users/**/Desktop/사과/labels"

# JSON 파일 변환 함수 호출
convert(json_dir)
```

## 좌표 변환 잘 됐는지 테스트 해보기
